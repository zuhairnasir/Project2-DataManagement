In this analysis, we aim to investigate which times of day Morning, Afternoon, Evening and we add another one time which is Night to experience the lowest average delays in flight arrivals and departures. We will categorize the dataset based on the time of day using the CRSDepTime (scheduled departure time) to group the flights into the three timeframes:

`Morning: 5:00 AM - 11:59 AM`

`Afternoon: 12:00 PM - 4:59 PM`

`Evening: 5:00 PM - 8:59 PM`

`Night: 9:00 PM - 4:59AM` 

We will exclude the Night timeframe from the analysis, as per the updated requirement. The two types of delays we will focus on are:

- Arrival Delays (ArrDelay): The delay in reaching the destination.

- Departure Delays (DepDelay): The delay in the scheduled departure.

Both delays are essential to understand the overall performance of flights, as departure delays often indicate problems with flight schedules, while arrival delays can be impacted by other external factors.

````python
query1 = '''
SELECT
  TimeOfDay,
  AVG(ArrDelay) AS AvgArrivalDelay,
  AVG(DepDelay) AS AvgDepartureDelay
FROM (
  SELECT *,
    CASE
      WHEN CRSDepTime BETWEEN 500 AND 1159 THEN 'Morning'
      WHEN CRSDepTime BETWEEN 1200 AND 1659 THEN 'Afternoon'
      WHEN CRSDepTime BETWEEN 1700 AND 2059 THEN 'Evening'
      WHEN (CRSDepTime >= 2100 AND CRSDepTime <= 2359) OR (CRSDepTime >= 0 AND CRSDepTime < 500) THEN 'Night'
      ELSE 'Unknown'
    END AS TimeOfDay
  FROM airlines2007
  WHERE Cancelled = 0
) AS categorized
GROUP BY TimeOfDay
ORDER BY AvgArrivalDelay ASC
'''

data_frame1 = pull_data(query1)
data_frame1
````

## 2. Delay Factors

### i. Identify and rank the top 3-5 factors contributing to flight delays, based on the delay categories provided in the dataset.

In this part of the analysis, we will identify the key factors contributing to flight delays, such as Carrier Delay, Weather Delay, NAS Delay, Security Delay, and Late Aircraft Delay. By summing the total delay time for each of these factors, we can rank them in terms of their total delay contribution.

This will provide insights into which factors are most responsible for delays, helping airlines and passengers understand the main contributors to flight disruptions.

````python
query4 = '''
SELECT
  DelayType,
  AVG(AvgDelayMinutes) AS AvgDelayMinutes,
  SUM(IncidentCount) AS IncidentCount
FROM (
  SELECT 'CarrierDelay' AS DelayType,
         AVG(CarrierDelay) AS AvgDelayMinutes,
         SUM(CASE WHEN CarrierDelay > 0 THEN 1 ELSE 0 END) AS IncidentCount
  FROM airlines2007
  WHERE CarrierDelay > 0

  UNION ALL

  SELECT 'WeatherDelay',
         AVG(WeatherDelay),
         SUM(CASE WHEN WeatherDelay > 0 THEN 1 ELSE 0 END)
  FROM airlines2007
  WHERE WeatherDelay > 0

  UNION ALL

  SELECT 'NASDelay',
         AVG(NASDelay),
         SUM(CASE WHEN NASDelay > 0 THEN 1 ELSE 0 END)
  FROM airlines2007
  WHERE NASDelay > 0

  UNION ALL

  SELECT 'SecurityDelay',
         AVG(SecurityDelay),
         SUM(CASE WHEN SecurityDelay > 0 THEN 1 ELSE 0 END)
  FROM airlines2007
  WHERE SecurityDelay > 0

  UNION ALL

  SELECT 'LateAircraftDelay',
         AVG(LateAircraftDelay),
         SUM(CASE WHEN LateAircraftDelay > 0 THEN 1 ELSE 0 END)
  FROM airlines2007
  WHERE LateAircraftDelay > 0
) AS DelaySummary
GROUP BY DelayType
ORDER BY AvgDelayMinutes DESC
LIMIT 5
'''

data_frame4 = pull_data(query4)
data_frame4
````

## 3. Cancellation Analysis

### i. Identify the primary reasons for flight cancellations as categorized in the dataset.

To gain deeper insight into flight performance, we can also explore cancellations. After examining the delay patterns and identifying key causes, the next logical step is to investigate cancellations. In this context, cancellations could be driven by various factors, such as operational issues, weather conditions, or external factors like air traffic control. From the dataset, the cancellation reasons are categorized similarly to delays, with the addition of a specific `Late Aircraft` category, which is typically associated with a cascade effect of delays from previous flights.

For our analysis, we will focus on the following cancellation categories:

+ `Weather-related cancellations`: Flights canceled due to weather conditions such as storms, poor visibility, or other adverse weather conditions.

+ `Carrier-related cancellations`: Cancellations stemming from issues within the airline's control, such as maintenance issues, crew shortages, or equipment failures.

+ `NAS-related cancellations`: Cancellations caused by air traffic control or other issues within the National Airspace System, such as congestion or restrictions.

````python
query7 = '''
SELECT * 
FROM (
    SELECT 
        d.UniqueCarrier AS CarrierCode,
        c.Description AS Description,
        SUM(CASE WHEN d.Cancelled > 0 THEN 1 ELSE 0 END) AS cancellation_count, 
        COUNT(*) AS total_flights,
        ROUND(100.0 * SUM(CASE WHEN d.Cancelled > 0 THEN 1 ELSE 0 END) / COUNT(*), 2) AS cancellation_rate_pct
    FROM 
        airlines2007 d
    LEFT JOIN 
        carriers c ON d.UniqueCarrier = c.Code
    GROUP BY 
        d.UniqueCarrier, c.Description
) sub
ORDER BY 
    cancellation_rate_pct DESC
'''

df7 = pull_data(query7)
df7
````

````python
# Group the top 5 airlines by cancellation rate
top5 = df7.head(5)
top5_airline = top5['description'].tolist()

# Create a new 'Group' column to classify the airlines into 'Top 5' and 'Others'
df7['Group'] = df7['description'].apply(lambda x: 'Top 5 Airlines' if x in top5_airline else 'Other Airlines')

# Create a boxplot for the top 5 airlines compared to the rest
plt.figure(figsize=(12, 10))
sns.boxplot(x='Group', y='cancellation_rate_pct', data=df7, palette='coolwarm')  # Changed color palette to 'coolwarm'
plt.title('Cancellation Rate Distribution for Top 5 vs Other Airlines', fontsize=16)  # Updated title for better readability
plt.ylabel('Cancellation Rate (%)', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adjusted gridline style
plt.tight_layout()
plt.show()
````

The **boxplot** above shows the distribution of **cancellation rates** between the **Top 5 Airlines** and **Other Airlines**. From the plot, we can see that the **Top 5 Airlines** have a higher **average cancellation rate** compared to **Other Airlines**. 

**Hypothesis:**
- **H0 (Null Hypothesis)**: There is **no significant difference** in the mean cancellation rates between the **Top 5 Airlines** and the **Other Airlines**.
- **H1 (Alternative Hypothesis)**: There is **significant difference** in the mean cancellation rates between the **Top 5 Airlines** and the **Other Airlines**.


````python
from scipy.stats import ttest_ind

# Filter the data for the top 5 and other airlines
top5_data = df7[df7['Group'] == 'Top 5 Airlines']['cancellation_rate_pct']
others_data = df7[df7['Group'] == 'Other Airlines']['cancellation_rate_pct']

# Perform an independent t-test to compare means
t_stat, p_value = ttest_ind(top5_data, others_data, equal_var=False)

# Print results with interpretation
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3f}")

# Interpretation
if p_value < 0.05:
    print("Result: The difference in cancellation rates between Top 5 Airlines and Other Airlines is statistically significant.")
else:
    print("Result: There is no statistically significant difference in cancellation rates between Top 5 Airlines and Other Airlines.")
````

Therefore, it can be concluded that the Top 5 Airlines experience a higher cancellation rate, which could be due to various operational factors, including maintenance issues, flight scheduling, or other internal airline challenges. These factors should be examined further to understand the specific reasons behind the higher cancellation rates.

````python
query7 = '''
WITH airport_stats AS (
    SELECT COALESCE(o.Origin, d.Dest) AS AirportCode, 
           o.cancellation_count AS Origin_Cancellation_count, 
           o.total_flights AS Origin_Total_Flights, 
           CASE WHEN o.total_flights > 0 THEN (o.cancellation_count * 100.0 / o.total_flights) ELSE 0 END AS Origin_Cancellation_Rate, 
           d.cancellation_count AS Dest_Cancellation_count, 
           d.total_flights AS Dest_Total_Flights, 
           CASE WHEN d.total_flights > 0 THEN (d.cancellation_count * 100.0 / d.total_flights) ELSE 0 END AS Dest_Cancellation_Rate
    FROM (
        SELECT Origin, 
               SUM(CASE WHEN Cancelled > 0 THEN 1 ELSE 0 END) AS cancellation_count, 
               COUNT(*) AS total_flights 
        FROM airlines2007 
        GROUP BY Origin
    ) o 
    FULL OUTER JOIN (
        SELECT Dest, 
               SUM(CASE WHEN Cancelled > 0 THEN 1 ELSE 0 END) AS cancellation_count, 
               COUNT(*) AS total_flights 
        FROM airlines2007 
        GROUP BY Dest
    ) d ON o.Origin = d.Dest
)

SELECT 
    a.AirportCode,
    ap.airport AS AirportName,
    ap.city AS City,
    ap.state AS State,
    ap.country AS Country,
    ap.lat AS Latitude,
    ap.long AS Longitude,
    a.Origin_Cancellation_count,
    a.Origin_Total_Flights,
    a.Origin_Cancellation_Rate,
    a.Dest_Cancellation_count,
    a.Dest_Total_Flights,
    a.Dest_Cancellation_Rate
FROM 
    airport_stats a
LEFT JOIN 
    airports ap ON a.AirportCode = ap.iata
ORDER BY 
    a.Origin_Cancellation_Rate DESC, 
    a.Dest_Cancellation_Rate DESC
'''

data_frame7 = pull_data(query7)
data_frame7.head()
````
